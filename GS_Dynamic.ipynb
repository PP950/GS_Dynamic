```

import pandas as pd
from typing import List, Tuple, Dict


# ============================================================
# 1) Construção do DataFrame com >= 20 trilhas de capacitação
# ============================================================

def construir_dataframe_trilhas() -> pd.DataFrame:
    """
    Cria um DataFrame com pelo menos 20 trilhas de capacitação
    relacionadas ao futuro do trabalho (IA, dados, soft skills, etc.).
    """
    trilhas = [
        {
            "id": 1,
            "nome": "IA aplicada ao trabalho",
            "area": "Tecnologia/IA",
            "horas": 12,
            "custo": 1200,
            "impacto_empregabilidade": 9.5,
            "ods_principal": 4
        },
        {
            "id": 2,
            "nome": "Fundamentos de análise de dados",
            "area": "Dados/Analytics",
            "horas": 10,
            "custo": 900,
            "impacto_empregabilidade": 9.2,
            "ods_principal": 9
        },
        {
            "id": 3,
            "nome": "Cibersegurança para profissionais não técnicos",
            "area": "Segurança",
            "horas": 8,
            "custo": 800,
            "impacto_empregabilidade": 8.5,
            "ods_principal": 9
        },
        {
            "id": 4,
            "nome": "Soft skills: comunicação e empatia em times híbridos",
            "area": "Soft Skills",
            "horas": 6,
            "custo": 600,
            "impacto_empregabilidade": 8.8,
            "ods_principal": 8
        },
        {
            "id": 5,
            "nome": "Liderança de equipes remotas",
            "area": "Gestão",
            "horas": 7,
            "custo": 750,
            "impacto_empregabilidade": 8.9,
            "ods_principal": 8
        },
        {
            "id": 6,
            "nome": "Design de experiências inclusivas",
            "area": "Inclusão",
            "horas": 5,
            "custo": 650,
            "impacto_empregabilidade": 8.3,
            "ods_principal": 10
        },
        {
            "id": 7,
            "nome": "Trabalho híbrido e produtividade",
            "area": "Gestão",
            "horas": 4,
            "custo": 500,
            "impacto_empregabilidade": 7.8,
            "ods_principal": 8
        },
        {
            "id": 8,
            "nome": "Ferramentas de colaboração digital",
            "area": "Tecnologia/IA",
            "horas": 5,
            "custo": 550,
            "impacto_empregabilidade": 7.9,
            "ods_principal": 9
        },
        {
            "id": 9,
            "nome": "Saúde mental e bem-estar no trabalho",
            "area": "Saúde e Bem-estar",
            "horas": 6,
            "custo": 700,
            "impacto_empregabilidade": 8.7,
            "ods_principal": 3
        },
        {
            "id": 10,
            "nome": "Economia verde e negócios sustentáveis",
            "area": "Sustentabilidade",
            "horas": 9,
            "custo": 950,
            "impacto_empregabilidade": 8.9,
            "ods_principal": 8
        },
        {
            "id": 11,
            "nome": "Ética e IA responsável",
            "area": "Tecnologia/IA",
            "horas": 5,
            "custo": 600,
            "impacto_empregabilidade": 8.6,
            "ods_principal": 10
        },
        {
            "id": 12,
            "nome": "Programação básica para automação de tarefas",
            "area": "Tecnologia/IA",
            "horas": 10,
            "custo": 1000,
            "impacto_empregabilidade": 9.0,
            "ods_principal": 4
        },
        {
            "id": 13,
            "nome": "Requalificação acelerada (reskilling) em dados",
            "area": "Dados/Analytics",
            "horas": 14,
            "custo": 1300,
            "impacto_empregabilidade": 9.4,
            "ods_principal": 8
        },
        {
            "id": 14,
            "nome": "Gamificação para engajamento em equipes distribuídas",
            "area": "Gestão",
            "horas": 6,
            "custo": 650,
            "impacto_empregabilidade": 8.1,
            "ods_principal": 9
        },
        {
            "id": 15,
            "nome": "Recrutamento inclusivo apoiado por dados",
            "area": "Inclusão",
            "horas": 7,
            "custo": 800,
            "impacto_empregabilidade": 8.4,
            "ods_principal": 10
        },
        {
            "id": 16,
            "nome": "Comunidades globais de aprendizagem",
            "area": "Educação",
            "horas": 5,
            "custo": 550,
            "impacto_empregabilidade": 8.0,
            "ods_principal": 4
        },
        {
            "id": 17,
            "nome": "Produtividade pessoal em regime híbrido",
            "area": "Soft Skills",
            "horas": 4,
            "custo": 450,
            "impacto_empregabilidade": 7.7,
            "ods_principal": 8
        },
        {
            "id": 18,
            "nome": "Introdução a plataformas de trabalho por projetos",
            "area": "Gestão",
            "horas": 5,
            "custo": 500,
            "impacto_empregabilidade": 7.9,
            "ods_principal": 8
        },
        {
            "id": 19,
            "nome": "Diversidade e redução de desigualdades no trabalho",
            "area": "Inclusão",
            "horas": 6,
            "custo": 700,
            "impacto_empregabilidade": 8.5,
            "ods_principal": 10
        },
        {
            "id": 20,
            "nome": "Infraestrutura digital e conectividade",
            "area": "Tecnologia/IA",
            "horas": 8,
            "custo": 900,
            "impacto_empregabilidade": 8.2,
            "ods_principal": 9
        },
        {
            "id": 21,
            "nome": "Carreira em economia de impacto social",
            "area": "Sustentabilidade",
            "horas": 9,
            "custo": 950,
            "impacto_empregabilidade": 8.8,
            "ods_principal": 8
        },
        {
            "id": 22,
            "nome": "Aprendizagem ao longo da vida (lifelong learning)",
            "area": "Educação",
            "horas": 6,
            "custo": 650,
            "impacto_empregabilidade": 8.9,
            "ods_principal": 4
        }
    ]

    df = pd.DataFrame(trilhas)
    return df


# ============================================================
# 2) Ordenação com MERGE SORT recursivo + memoização
# ============================================================

def _merge_ids(left_ids: List[int],
               right_ids: List[int],
               df: pd.DataFrame,
               col: str) -> List[int]:
    """
    Função auxiliar que faz o merge de duas listas de ids já ordenadas
    com base na coluna `col` do DataFrame.
    """
    result = []
    i = j = 0

    while i < len(left_ids) and j < len(right_ids):
        if df.loc[df["id"] == left_ids[i], col].values[0] >= df.loc[df["id"] == right_ids[j], col].values[0]:
            result.append(left_ids[i])
            i += 1
        else:
            result.append(right_ids[j])
            j += 1

    # Adiciona o restante
    result.extend(left_ids[i:])
    result.extend(right_ids[j:])
    return result


def merge_sort_ids(ids: Tuple[int, ...],
                   df: pd.DataFrame,
                   col: str,
                   memo: Dict[Tuple[int, ...], List[int]]) -> List[int]:
    """
    Merge sort recursivo sobre uma tupla de ids, usando memoização.
    Retorna uma lista de ids ordenada pela coluna `col` do DataFrame
    em ordem decrescente (maior impacto primeiro).
    """
    if ids in memo:
        return memo[ids]

    if len(ids) <= 1:
        memo[ids] = list(ids)
        return memo[ids]

    mid = len(ids) // 2
    left_ids = ids[:mid]
    right_ids = ids[mid:]

    sorted_left = merge_sort_ids(left_ids, df, col, memo)
    sorted_right = merge_sort_ids(right_ids, df, col, memo)

    merged = _merge_ids(sorted_left, sorted_right, df, col)
    memo[ids] = merged
    return merged


def ordenar_trilhas_por_impacto(df: pd.DataFrame, col: str = "impacto_empregabilidade") -> pd.DataFrame:
    """
    Ordena o DataFrame de trilhas de capacitação usando merge sort recursivo
    com memoização sobre os ids, baseado na coluna `col`.
    """
    ids_tuple = tuple(df["id"].tolist())
    memo = {}
    sorted_ids = merge_sort_ids(ids_tuple, df, col, memo)

    df_sorted = df.set_index("id").loc[sorted_ids].reset_index()
    return df_sorted


# ============================================================
# 3) Mochila 0/1 recursiva + memoização (max impact, limite horas)
# ============================================================

def mochila_recursiva(index: int,
                      horas_restantes: int,
                      itens: List[Dict],
                      memo: Dict[Tuple[int, int], Tuple[float, List[int]]]
                      ) -> Tuple[float, List[int]]:
    """
    Resolve o problema da mochila 0/1 usando recursão e memoização.
    - index: índice do item atual
    - horas_restantes: capacidade de horas disponível
    - itens: lista de dicionários com 'id', 'horas', 'impacto_empregabilidade'
    Retorna uma tupla (impacto_total, lista_ids_escolhidos).
    """
    chave = (index, horas_restantes)
    if chave in memo:
        return memo[chave]

    if index == len(itens) or horas_restantes == 0:
        memo[chave] = (0.0, [])
        return memo[chave]

    item = itens[index]

    # Opção 1: não pegar o item atual
    impacto_sem, ids_sem = mochila_recursiva(index + 1, horas_restantes, itens, memo)

    # Opção 2: pegar o item atual (se couber)
    if item["horas"] > horas_restantes:
        melhor = (impacto_sem, ids_sem)
    else:
        impacto_com, ids_com = mochila_recursiva(
            index + 1,
            horas_restantes - item["horas"],
            itens,
            memo
        )
        impacto_com += item["impacto_empregabilidade"]
        ids_com = ids_com + [item["id"]]

        # Escolhe a melhor opção
        if impacto_com > impacto_sem:
            melhor = (impacto_com, ids_com)
        else:
            melhor = (impacto_sem, ids_sem)

    memo[chave] = melhor
    return melhor


def selecionar_trilhas_otimas(df: pd.DataFrame, max_horas: int) -> Tuple[float, List[int]]:
    """
    Prepara os dados e chama a mochila recursiva para selecionar
    as trilhas com maior impacto de empregabilidade dentro do limite de horas.
    """
    itens = df[["id", "horas", "impacto_empregabilidade"]].to_dict(orient="records")
    memo = {}
    melhor_impacto, lista_ids = mochila_recursiva(0, max_horas, itens, memo)
    return melhor_impacto, lista_ids


# ============================================================
# 4) Funções recursivas auxiliares para o relatório (com memo)
# ============================================================

def soma_metricas_recursiva(ids: Tuple[int, ...],
                            df: pd.DataFrame,
                            col: str,
                            memo: Dict[Tuple[int, str], float]) -> float:
    """
    Soma recursivamente um atributo numérico (col) para um conjunto de ids,
    usando memoização.
    """
    chave = (ids, col)
    if chave in memo:
        return memo[chave]

    if not ids:
        memo[chave] = 0.0
        return memo[chave]

    # divide e conquista: soma do último + soma do resto
    ultimo = ids[-1]
    resto = ids[:-1]

    valor_ultimo = float(df.loc[df["id"] == ultimo, col].values[0])
    soma_resto = soma_metricas_recursiva(resto, df, col, memo)

    total = valor_ultimo + soma_resto
    memo[chave] = total
    return total


def gerar_relatorio(df: pd.DataFrame, ids_escolhidos: List[int]) -> str:
    """
    Gera um relatório textual com base nas trilhas escolhidas pela mochila.
    Usa soma recursiva/memoizada para totalizar horas e impacto.
    """
    memo_somas = {}
    ids_tuple = tuple(ids_escolhidos)

    total_horas = soma_metricas_recursiva(ids_tuple, df, "horas", memo_somas)
    total_impacto = soma_metricas_recursiva(ids_tuple, df, "impacto_empregabilidade", memo_somas)

    trilhas_escolhidas = df.set_index("id").loc[list(ids_escolhidos)].reset_index()

    linhas = []
    linhas.append("===== RELATÓRIO DE CAPACITAÇÃO PARA O FUTURO DO TRABALHO =====")
    linhas.append("")
    linhas.append(f"Quantidade de trilhas escolhidas: {len(ids_escolhidos)}")
    linhas.append(f"Total de horas: {total_horas:.1f} h")
    linhas.append(f"Impacto total estimado na empregabilidade: {total_impacto:.2f}")
    linhas.append("")
    linhas.append("Trilhas selecionadas (ordenadas por impacto):")

    trilhas_ordenadas = trilhas_escolhidas.sort_values(by="impacto_empregabilidade", ascending=False)

    for _, row in trilhas_ordenadas.iterrows():
        linhas.append(
            f"- [{row['id']:02d}] {row['nome']} "
            f"| Área: {row['area']} "
            f"| Horas: {row['horas']} "
            f"| Impacto: {row['impacto_empregabilidade']:.1f} "
            f"| ODS principal: {row['ods_principal']}"
        )

    # Resumo por área (usando groupby normal para não exagerar na recursão)
    linhas.append("")
    linhas.append("Resumo por área (trilhas selecionadas):")
    resumo_area = trilhas_escolhidas.groupby("area").agg(
        horas_totais=("horas", "sum"),
        impacto_medio=("impacto_empregabilidade", "mean"),
        qtd_trilhas=("id", "count")
    ).reset_index()

    for _, row in resumo_area.iterrows():
        linhas.append(
            f"- Área: {row['area']} | Trilhas: {row['qtd_trilhas']} "
            f"| Horas totais: {row['horas_totais']} "
            f"| Impacto médio: {row['impacto_medio']:.2f}"
        )

    return "\n".join(linhas)


# ============================================================
# 5) Execução de exemplo (NÃO usar só 5 ou 6 infos!)
# ============================================================

if __name__ == "__main__":
    # 1) Construir DataFrame com 22 trilhas (>= 20, como pedido)
    df_trilhas = construir_dataframe_trilhas()

    print("===== TRILHAS ORIGINAIS (AMOSTRA) =====")
    print(df_trilhas.head(10))
    print("\nTotal de trilhas:", len(df_trilhas))

    # 2) Ordenar trilhas por impacto usando merge sort + memoização
    df_ordenado = ordenar_trilhas_por_impacto(df_trilhas, col="impacto_empregabilidade")

    print("\n===== TRILHAS ORDENADAS POR IMPACTO (TOP 10) =====")
    print(df_ordenado.head(10)[["id", "nome", "impacto_empregabilidade", "horas"]])

    # 3) Definir limite de horas de estudo (capacidade da "mochila")
    max_horas = 40  # por exemplo, 40 horas disponíveis

    # 4) Resolver o problema da mochila 0/1
    melhor_impacto, ids_escolhidos = selecionar_trilhas_otimas(df_trilhas, max_horas)

    print("\n===== RESULTADO MOCHILA =====")
    print(f"Impacto total: {melhor_impacto:.2f}")
    print(f"IDs escolhidos: {sorted(ids_escolhidos)}")

    # 5) Gerar relatório completo
    relatorio = gerar_relatorio(df_trilhas, ids_escolhidos)
    print("\n" + relatorio)
```
